```python
import os 
os.makedirs("../data/text_files",exist_ok=True)
```

```python
sample_text = {
    "../data/text_files/sample1.txt": "This is the content of sample1.txt. It contains some text data for testing.",
}

for filepath, content in sample_text.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)

print("done creating text file")
```

```python
# Document Loader
#from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/sample1.txt")
document = loader.load()
print(document);
```

```python
# Directory Loader
from langchain_community.document_loaders import DirectoryLoader
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt", # File pattern
    loader_cls=TextLoader, #Loader to use
    loader_kwargs={"encoding": "utf-8"},
    show_progress=False
)
documents = dir_loader.load()
documents
```

```python
# PDF Loaders
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls=PyPDFLoader,
    show_progress=False
)
pdf_documents = dir_loader.load()
pdf_documents
```
